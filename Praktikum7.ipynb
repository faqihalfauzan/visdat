**Import Library**


```python
import numpy as np
import matplotlib.pyplot as plt
```

**Dataset**

```python
#data Penjualan
brands = ['Samsung', 'Iphone', 'Vivo', 'Oppo']
sales_2023 =[500, 700, 300, 400]
sales_2024 =[550, 750, 350, 450]
sales_2025 =[600, 800, 400, 500]
```

**Basic Chart (Horizontal & Stracked Horizontal Bar Chart**

```python
y_pos = np.arange(len(brands))

# Canvas dan axis grafik
fig1, ax1, = plt.subplots()

# Grafik Batang Horizontal
ax1.set_title('Horizontal Bar Chart - 2023')
ax1.set_xlabel('Jumlah Penjualan')
ax1.set_ylabel('Merk')
ax1.barh(brands, sales_2023)
plt.show()

# stacked horizontal bar chart
fig2, ax2 = plt.subplots()# canvas dan axis grafik

# grafik batang horizontal
ax2.set_title('Stacked Horizontal Bar Chart - 2023  & 2024')
ax2.set_xlabel('Jumlah Penjualan')
ax2.set_ylabel('Merk')

# bar pertama untuk 2023
ax2.barh(brands, sales_2023, label='2023')

# bar kedua untuk 2024
ax2.barh(brands, sales_2024, left=sales_2023,  label='2024')

# legenda
ax2.legend()
plt.show()
```

**Kustomisasi Grafik**

```python
# Data (Asumsi)
brands = ['Samsung', 'Apple', 'Xiaomi', 'Oppo']
sales_2023 = [1200, 1100, 900, 800]
sales_2024 = [1300, 1150, 950, 850]
colors = ['gold', 'skyblue', 'coral', 'lightgreen']

# --- Custom Horizontal Bar Chart (2023) ---
fig3, ax3 = plt.subplots()

ax3.barh(brands, sales_2023, color=colors, edgecolor='black', height=0.6)
ax3.set_title('Penjualan Smartphone Tahun 2023 (Custom)')
ax3.set_xlabel('Jumlah Penjualan')
ax3.set_ylabel('Merk')
ax3.grid(axis='x', linestyle='--', alpha=0.7)

# Menampilkan label nilai
for i, v in enumerate(sales_2023):
    ax3.text(v + 10, i, str(v), color='black', fontweight='bold', va='center')

plt.show()

# --- Custom Stacked Horizontal Bar Chart (2023-2024) ---
fig4, ax4 = plt.subplots()

ax4.barh(brands, sales_2023, color='lightcoral', edgecolor='black', label='2023')
ax4.barh(brands, sales_2024, left=sales_2023, color='lightskyblue', edgecolor='black', label='2024')

ax4.set_title('Penjualan Smartphone Tahun 2023 & 2024 (Custom)')
ax4.set_xlabel('Jumlah Penjualan')
ax4.set_ylabel('Merk')
ax4.grid(axis='x', linestyle='--', alpha=0.5)

# Hitung total penjualan untuk label
total_sales = np.array(sales_2023) + np.array(sales_2024)

# Label nilai total di ujung bar
for i, v in enumerate(total_sales):
    ax4.text(v + 10, i, str(v), va='center', fontweight='bold')

# PERBAIKAN: Tambahkan legend dan show
ax4.legend()
plt.show()
```

**Multiple Cart**

```python
# --- 1. Multiple Horizontal Bar Chart (Grouped) ---
fig5, ax5 = plt.subplots()

# bar pertama
ax5.barh(y_pos - 0.2, sales_2023, color='skyblue', height=0.4, label='2023')
# bar kedua
ax5.barh(y_pos + 0.2, sales_2024, color='salmon', height=0.4, label='2024')

ax5.set_yticks(y_pos, labels=brands)
ax5.set_title('Penjualan Smartphone Tahun 2023 dan 2024 (Multiple Chart)')
ax5.set_xlabel('Jumlah Penjualan')
ax5.set_ylabel('Merk')
ax5.legend()

plt.show()

# --- 2. Multiple Stacked Horizontal Bar Chart ---
fig6, ax6 = plt.subplots()

# bar pertama
ax6.barh(y_pos, sales_2023, color='skyblue', label='2023')
# bar kedua
ax6.barh(y_pos, sales_2024, left=sales_2023, color='salmon', label='2024')
# bar ketiga (Total Stacked)
ax6.barh(y_pos, sales_2025, left=np.array(sales_2023) + np.array(sales_2024), color='lightgreen', label='2025')

ax6.set_yticks(y_pos, labels=brands)
ax6.set_title('Penjualan Smartphone Tahun 2023 vs 2024 vs 2025 (Stacked)')
ax6.set_xlabel('Jumlah Penjualan')
ax6.set_ylabel('Merk')
ax6.legend()

plt.show()
```

**Pie Chart**


```python
import matplotlib.pyplot as plt
import numpy as np

# @title Pilih Kategori Visualisasi { run: "auto" }
kategori = "Multiple Chart" # @param ["Basic Chart", "Kustomisasi Grafik", "Multiple Chart"]

# DATASET
kegiatan = ['Olahraga', 'Seni', 'Musik', 'Debat']
persentase = [30, 20, 25, 25]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep']
shoes = [500, 600, 700, 800, 650, 700, 750, 800, 850]
sandals = [300, 350, 400, 450, 500, 480, 520, 550, 600]
socks = [200, 250, 300, 350, 400, 380, 420, 450, 500]

# LOGIKA VISUALISASI
if kategori == "Basic Chart":
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    ax1.pie(persentase, labels=kegiatan, autopct='%1.1f%%', startangle=140)
    ax1.set_title("Basic Pie Chart")

    ax2.fill_between(months, shoes, alpha=0.4)
    ax2.set_title("Basic Area Chart")
    plt.show()

elif kategori == "Kustomisasi Grafik":
    colors = ['gold', 'lightskyblue', 'lightcoral', 'lightgreen']
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    ax1.pie(persentase, labels=kegiatan, colors=colors, autopct='%1.1f%%', explode=(0.1, 0, 0, 0), shadow=True)
    ax1.set_title("Customized Pie Chart")

    ax2.fill_between(months, shoes, alpha=0.5, label='Shoes', color='skyblue')
    ax2.fill_between(months, sandals, alpha=0.5, label='Sandals', color='orange')
    ax2.legend()
    ax2.grid(axis='y', linestyle='--')
    ax2.set_title("Customized Area Chart")
    plt.show()

else:
    fig, axs = plt.subplots(2, 1, figsize=(10, 10))
    axs[0].pie([40, 60], labels=['Grup A', 'Grup B'], autopct='%1.1f%%')
    axs[0].set_title("Multiple Chart - Pie")

    axs[1].fill_between(months, shoes, alpha=0.4, label='Shoes')
    axs[1].fill_between(months, sandals, alpha=0.4, label='Sandals')
    axs[1].fill_between(months, socks, alpha=0.4, label='Socks')
    axs[1].legend()
    axs[1].set_title("Multiple Chart - Area")
    plt.tight_layout()
    plt.show()
```
