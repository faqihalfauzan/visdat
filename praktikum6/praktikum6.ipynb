#### **1. Import Library**

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### **2. Dataset**

```python
# datasaet
stores = ['Store A', 'Store B', 'Store C']
product_a = [200, 250, 300]
product_b = [150, 300, 200]

# data quarter
q1_male = [150, 180, 160]
q1_female = [140, 200, 180]
q2_male = [170, 190, 175]
q2_female = [130, 210, 160]
q3_male = [160, 185, 170]
q3_female = [135, 195, 165]
q4_male = [155, 175, 168]
q4_female = [130, 210, 160]

male = q1_male
female = q1_female
```

#### **3. Basic Stacked Vertical Bar Chart**

```python
x = np.arange(len(stores))
plt.bar(x, male, label='Male', color='blue')
plt.bar(x, female, bottom=male, label='Female', color='pink')

plt.title('Population by Gender and Store')
plt.xlabel('Stores')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend()

plt.show()
```

#### **4. Kustomisasi Stacked Vertical Bar Chart**

```python
x = np.arange(len(stores))
plt.bar(x, male, label='Male', color='blue')
plt.bar(x, female, bottom=male, label='Female', color='pink')

# anotasi bar
for i in range(len(x)):
    plt.text(x[i], male[i]/2, str(male[i]), ha='center', va='center', color='white')
    plt.text(x[i], male[i] + female[i]/2, str(female[i]), ha='center', va='center', color='black')

plt.title('Population by Gender and Store')
plt.xlabel('Stores')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend()

plt.show()
```

#### **5. Multiple Stacked Vertical Bar Chart**

```python
width = 0.15
x = np.arange(len(stores))

# quarter 1
plt.bar(x - width*1.5, q1_male, label='Q1 Male', color='lightblue', width=width)
plt.bar(x - width*1.5, q1_female, bottom=q1_male, label='Q1 Female', color='pink', width=width)

# quarter 2
plt.bar(x - width/2, q2_male, label='Q2 Male', color='blue', width=width)
plt.bar(x - width/2, q2_female, bottom=q2_male, label='Q2 Female', color='red', width=width)

# quarter 3
plt.bar(x + width/2, [160,185,170], label='Q3 Male', color='green', width=width)
plt.bar(x + width/2, [135,195,165], bottom=[160,185,170], label='Q3 Female', color='yellow', width=width)

# quarter 4
plt.bar(x + width*1.5, [155,175,168], label='Q4 Male', color='purple', width=width)
plt.bar(x + width*1.5, [130,210,160], bottom=[155,175,168], label='Q4 Female', color='orange', width=width)

plt.title('Population by Gender and Store (Multiple Quarters)')
plt.xlabel('Stores')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
