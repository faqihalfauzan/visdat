```python
import matplotlib.pyplot as plt
```

```python
# membuat data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Des']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]
```

**Single Line Plot**

```python
# line plot 1 garis, style grafik deafult
plt.plot(months, product_A_sales, label='Product A')
plt.title('Penjualan Produk A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# kustomisasi line plot (line 1 -product - A)
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker='o', color='green')
# kustomisasi line plot (line 1 -product - B)
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker='x', color='red')

plt.title('Penjualan Produk A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Jenis Garis untuk Menunjukan Tren**



```python
# Data sample tambahan
product_C_sales = [18,21,22,23,14,55,67,53,63,55,77,88,99,98,76][:12]
product_D_sales = [31,21,33,43,32,55,36,71,12,43,88,97,64,43,76][:12]

#plot produk a
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker='o', color='green')

#plot produk b
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker='x', color='red')

#plot produk C
plt.plot(months, product_C_sales, label='Product C', linestyle=":", color='black')

#plot produk D
plt.plot(months, product_D_sales, label='Product D', linestyle='-.', color='blue')


plt.title('Penjualan Produk  per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Subplot pada Line Chart**

```python
# subplot, 2 baris 1 kolom, panjang 10 inch, tinggi 8 inch
fig, axs = plt.subplots(2, 1, figsize=(10,8))

# plot pertama untuk product C
axs[0].plot(months, product_C_sales, label='Product C', linestyle='--' ,color='green', marker='d')
axs[0].set_title('Product C')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

# plot kedua untuk product D
axs[1].plot(months, product_D_sales, label='Product D', color='blue', marker='o')
axs[1].set_title('Product D')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
