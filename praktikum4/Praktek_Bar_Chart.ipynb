```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Science'
]

jumlah_mahasiswa = [120, 150, 100, 80]


#Membuat basic bar chart
plt.bar(jurusan, jumlah_mahasiswa, color =['blue', 'orange', 'green', 'red'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa per Jurusan')
plt.show()

```

**a. Basic Bar Chart dengan Matplotlib**

```python
#Data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Science'
]

jumlah_mahasiswa = [120, 150, 100, 80]


#Membuat basic bar chart
plt.bar(jurusan, jumlah_mahasiswa, color =['blue', 'black', 'green', 'red'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa per Jurusan')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**b. Kostumisasi Basic Bart Chart**

```python
# Data
jurusan = ['Ilmu Komputer', 'Sistem Informasi', 'Teknik Informatika', 'Data Science']
jumlah_mahasiswa = [120, 150, 100, 80]

# Menambahkan kustomisasi
colors = ['blue', 'green', 'red', 'purple']
plt.bar(jurusan, jumlah_mahasiswa, color=colors)
plt.title('Jumlah Mahasiswa per Jurusan')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)

for i, v in enumerate(jumlah_mahasiswa):
    plt.text(i, v + 5, str(v), ha='center') # Menambahkan Label pada batang

plt.show()
```

**c. Menambahkan Multiple Bar Chart**

```python
# Data
jurusan = ['Ilmu Komputer', 'Sistem Informasi', 'Teknik Informatika', 'Data Science']
jumlah_mahasiswa = [120, 150, 100, 80]

# Data tambahan untuk tahun yang berbeda
jumlah_mahasiswa_2024 = [110, 140, 95, 85]

x = range(len(jurusan)) # Indeks posisi untuk batang
width = 0.4

# Batang untuk tahun 2023 (diletakkan di posisi x)
plt.bar(x, jumlah_mahasiswa, width, label='2023', color='skyblue')

# Batang untuk tahun 2024 (diletakkan sedikit bergeser dari x)
plt.bar([p + width for p in x], jumlah_mahasiswa_2024, width, label='2024', color='orange')

plt.title('Jumlah Mahasiswa per Jurusan (2023 vs 2024)')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)

# Mengatur Label x-axis di tengah antara dua batang
plt.xticks([p + width / 2 for p in x], jurusan)
plt.legend()
plt.show()
```

**d. Basic Bar Chart untuk Melihat Pola Tren**

```python
# Data Jumlah mahasiswa per tahun
jurusan = ['Ilmu Komputer', 'Sistem Informasi', 'Teknik Informatika', 'Data Science']
tahun = ['2019', '2020', '2021', '2022', '2023']
data = {
    'Ilmu Komputer': [100, 110, 120, 130, 140],
    'Sistem Informasi': [120, 125, 135, 145, 160],
    'Teknik Informatika': [90, 95, 100, 105, 110],
    'Data Science': [70, 75, 80, 85, 90]
}

# Membuat Bar Chart untuk masing-masing jurusan
width = 0.2
x = range(len(tahun))

# Visualisasi untuk masing-masing jurusan
# Jurusan 1: Ilmu Komputer (Posisi Awal: x)
plt.bar(x, data['Ilmu Komputer'], width=width, label='Ilmu Komputer', color='blue')

# Jurusan 2: Sistem Informasi (Digeser 1 * width)
plt.bar([p + width for p in x], data['Sistem Informasi'], width=width, label='Sistem Informasi', color='green')

# Jurusan 3: Teknik Informatika (Digeser 2 * width)
plt.bar([p + width * 2 for p in x], data['Teknik Informatika'], width=width, label='Teknik Informatika', color='orange')

# Jurusan 4: Data Science (Digeser 3 * width)
plt.bar([p + width * 3 for p in x], data['Data Science'], width=width, label='Data Science', color='purple')

# Memberikan detail pada grafik
plt.title('Tren Jumlah Mahasiswa per Jurusan (2019-2023)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Mahasiswa')

# Mengatur label X-axis di tengah kelompok batang
# Posisi Tengah = Posisi Awal (p) + (Lebar Total Kelompok / 2)
# Lebar Total Kelompok = 4 * width
plt.xticks([p + width * 1.5 for p in x], tahun)
plt.legend()

# Menampilkan grafik
plt.show()
```
